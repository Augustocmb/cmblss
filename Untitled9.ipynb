```python
camb_path = os.path.realpath(os.path.join(os.getcwd(),'..'))
sys.path.insert(0,camb_path)
import camb
from camb import model, initialpower, correlations
print('Using CAMB %s installed at %s'%(camb.__version__,os.path.dirname(camb.__file__)))
```

```python
#Calculate results for specified parameters and return CAMBdata instance for getting results.
pars = camb.CAMBparams()
pars.set_cosmology(H0=67.66, ombh2=0.02242, omch2=0.11933, mnu=0.06, omk=0.001, tau=0.0561)
pars.InitPower.set_params(As=2.105e-9, ns=0.9665, r=0)
pars.set_for_lmax(2500, lens_potential_accuracy=0)
```

```python
#calculate results for these parameters
results = camb.get_results(pars)
```

```python
#Set all CAMB parameters at once, including parameters which are part of the CAMBparams structure, as well as global parameters.
powers =results.get_cmb_power_spectra(pars, CMB_unit='muK')
for name in powers: print(name)
```

```python
totCL=powers['total']
unlensedCL=powers['unlensed_scalar']
print(totCL.shape)
#Python CL arrays are all zero based (starting at L=0), Note L=0,1 entries will be zero by default.
#The different CL are always in the order TT, EE, BB, TE (with BB=0 for unlensed scalar results).
ls = np.arange(totCL.shape[0])
```

```python
len(ls)
```

```python
plt.plot(ls,totCL[:,0], label=' theoretical CMB Power Spectrum by (Lewis et al 2006, Planck mission 2018) $')
plt.ylabel("$\dfrac{\ell(\ell+1)}{2\pi} C_\ell~[\mu K^2]$")
plt.xlabel("$\ell$")
plt.legend()
plt.xlim([50, 2500])
```

```python
# I generate the alms
alm=hp.sphtfunc.synalm(totCL[:,0], lmax=len(ls), new=True)
```

```python
high_nside=2048   #resolution of the maps
cmb_map = hp.alm2map(alm, nside=high_nside, lmax=len(ls))
```

```python
hp.mollview(cmb_map,min=-300*1e-6, max=300*1e-6, unit="K", title="CMB Temperature")
```
