```python
# Creating CMB maps with synfast
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import sys, platform, os
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import camb
from camb import model, initialpower
print('Using CAMB %s installed at %s'%(camb.__version__,os.path.dirname(camb.__file__)))
# make sure the version and path is what you expect
```

```python
#Set up a new set of parameters for CAMB
#The defaults give one massive neutrino and helium set using BBN consistency
pars = camb.set_params(H0=67.5, ombh2=0.022, omch2=0.122, mnu=0.06, omk=0, tau=0.06,  
                       As=2e-9, ns=0.965, halofit_version='mead', lmax=3000)
```

```python
#calculate results for these parameters
results = camb.get_results(pars)
```

```python
#get dictionary of CAMB power spectra
powers =results.get_cmb_power_spectra(pars, CMB_unit='muK')
for name in powers: print(name)
    
```

```python
#plot the total lensed CMB power spectra versus unlensed, and fractional difference
totCL=powers['total']
unlensedCL=powers['unlensed_scalar']
print(totCL.shape)
#Python CL arrays are all zero based (starting at L=0), Note L=0,1 entries will be zero by default.
```

```python
%load_ext memory_profiler

from memory_profiler import profile

@profile
def my_map():
    #Create a map(s) from cl(s).
    cmb_map=hp.synfast(totCL[:,0],2**20)
    return cmb_map

# Run the function with memory profiling
%mprun -f my_map my_map()
```

```python
# Resolution map used
a = [512, 1024, 2048]
# Storing memory used (GB)
b = [0.503, 0.629, 0.954]

plt.scatter(a, b, marker='*', color='blue')
plt.xlabel('NSIDE')
plt.ylabel('Stored Memory (GB)')
plt.title('Resolution Map vs. Stored Memory of a CMB map')  # Adding title
plt.grid(True)
plt.show()
```

```python
# Resolution map used
a = [512, 1024, 2048]
# Storing memory used (GB)
b = [0.503, 0.629, 0.954]

# Perform linear regression
slope, intercept = np.polyfit(a, b, 1)

# Create regression line points
x_regression = np.array([min(a), max(a)])
y_regression = slope * x_regression + intercept

# Plot scatter plot and regression line
plt.scatter(a, b, marker='*', color='blue', label='Data Points')
plt.plot(x_regression, y_regression, color='red', label='regresion')
plt.xlabel('NSIDE')
plt.ylabel('Stored Memory (GB)')
plt.title('Resolution Map vs. Stored Memory with Linear Regression')
plt.xticks(a)  # Set x-axis ticks to [512, 1024, 2048]
plt.grid(True)
plt.legend()
plt.show()
```
