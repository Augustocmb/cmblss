```python
import pickle

# Open and load the pickle file
with open('remastered_curves.p', 'rb') as file:
    remastered_data = pickle.load(file)

# Display the content
print(remastered_data)
```

```python
# Taking the real part
Cls=np.real(remastered)
unit_conv=10**12
ells_max=42
els=np.arange(2,43)
els
```

```python
import matplotlib.pyplot as plt

# Increase the figure size for better visibility
plt.figure(figsize=(8, 6))

# Plot with a thicker line for clarity
plt.plot(els, unit_conv * Cls * els * (els + 1) / (2 * np.pi), label=r'$Pseudo-C_\ell$', color='darkblue', linewidth=2)

# Add grid for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Add a legend and set font size
plt.legend(fontsize=12)

# Customize axis labels with larger font size
plt.xlabel(r'$\ell$', fontsize=14)
plt.ylabel(r'$\langle \tilde{D}_{\ell} \rangle $ [$\mu \mathrm{K}^2$]', fontsize=14)

# Use logarithmic scale for y-axis
plt.yscale('log')

# Adjust tick label size
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Display the plot
plt.show()
```
