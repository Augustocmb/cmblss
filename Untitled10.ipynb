# In this notebook I'll get the memory vs NSIDE plot for analize the time and memory usage.

```python
# Importing libraries
import healpy as hp
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read the map
Planck_map=hp.read_map('/home/akozameh/COM_CMB_IQU-smica_2048_R3.00_hm2.fits')
# Read the mask

masc=hp.read_map('/home/akozameh/COM_Mask_CMB-common-Mask-Int_2048_R3.00.fits')
```

```python
hp.mollview(masc)
```

```python
# Degrade the masp
masc_1=hp.pixelfunc.ud_grade(masc,16) 
# Degrade the map
Planck_map=hp.pixelfunc.ud_grade(Planck_map,16)
```

```python
# Saving files
# Saving the mask and map
hp.fitsfunc.write_map('/home/akozameh/reMASTERed/masc_1.fits', masc_1, overwrite=True)
hp.fitsfunc.write_map('/home/akozameh/reMASTERed/Planck_map.fits',Planck_map, overwrite= True)
```

```python
# Creating arrays of computing time
time=np.array([2.144860029220581,475.83250093460083,23149.113530158997])   #Computing time.
memory=np.array([0.252, 5,64])                                             #Memory usage in Gb.
#Resolution of the maps
nside=np.array([8,16,32])
```

```python
#Ploting the results
plt.scatter(nside,time, label='Computing time(Sersic)')
#plt.scatter(nside, memory, label='memory usage')
plt.legend()
plt.ylabel('Time(s)')
plt.xlabel('Resolution')
```

```python
#Ploting the results
plt.scatter(nside,memory, color='red', label='Memory usage(Sersic)')
#plt.scatter(nside, memory, label='memory usage')
plt.legend()
plt.ylabel('Memory (GB)')
plt.xlabel('Resolution')
```

```python
fig, ax = plt.subplots(2)

ax[0].scatter(nside, memory, color='red')
ax[0].set_title('Memory usage in function of the map resolution')
ax[0].set_ylabel('Memory (GB)')

ax[1].scatter(nside, time, color='blue')
ax[1].set_title('Computing time in Sersic')
ax[1].set_ylabel('Time (s)')

# Increase the separation between subplots
plt.subplots_adjust(hspace=0.5)


# Common x-axis label for both subplots
fig.text(0.5, 0.04, 'Resolution of the map', ha='center')

plt.show()
```

```python
# Interpolation of the curves.
from scipy import interpolate
f_1 = interpolate.interp1d(nside, memory, kind='quadratic')
f_2 = interpolate.interp1d(nside, time, kind='quadratic')
res_new=np.array([8,16,32])
# My new curves

y_1=f_1(res_new)
y_2=f_2(res_new)
```

```python
plt.plot(res_new,y_1)
```

```python
# Plotting the results
fig, ax = plt.subplots(2)

ax[0].plot(res_new, y_1, color='red')
ax[0].set_title('Quadratic Interpolation of the Memory usage in function of the map resolution')
ax[0].set_ylabel('Memory (GB)')

ax[1].plot(res_new, y_2, color='blue')
ax[1].set_title(' Qudratic Interpolation of Computing time in Sersic')
ax[1].set_ylabel('Time (s)')

# Increase the separation between subplots
plt.subplots_adjust(hspace=0.5)


# Common x-axis label for both subplots
fig.text(0.5, 0.5, 'Resolution of the map', ha='center')

plt.show()
```

```python
# Extrapolation the data until nside=128
from scipy.optimize import curve_fit
#Define a new function to fit the data.
x=np.array([3,4,5])
def func(a,x):
    return a*(2**x)
# Curve fitting
popt_1, pcov_1 = curve_fit(func, x, memory) # For memory usage
popt_2, pcov_2 = curve_fit(func, x, time) # For computing time

# Extrapolate beyond the available data
x_extrapolated = np.array([8,16,32,64,128])
y_extrapolated_1 = func(x_extrapolated, *popt_1)
y_extrapolated_2 = func(x_extrapolated, *popt_2)
```

```python
plt.scatter(nside, memory, label='Measured')
plt.plot(x_extrapolated, y_extrapolated_1,label='Extrapolation')
```

```python
# Plotting the results
fig, ax = plt.subplots(2)

ax[0].plot(x_extrapolated, y_extrapolated_1, color='red')
ax[0].set_title('Quadratic extrapolation of the Memory usage until nside=128')
ax[0].set_ylabel('Memory (GB)')


ax[1].plot(x_extrapolated, y_extrapolated_2, color='blue')
ax[1].set_title(' Qudratic extrapolation of Computing time until nside=128')
ax[1].set_ylabel('Time (s)')

# Increase the separation between subplots
plt.subplots_adjust(hspace=0.5)


# Common x-axis label for both subplots
fig.text(0.5, 0.5, 'Resolution of the map', ha='center')

plt.show()
```
