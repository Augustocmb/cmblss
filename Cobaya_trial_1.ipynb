```python
import numpy as np
from scipy import stats
```

```python
# Baseline function for perfoming likelihood

def gaussian_2d(x, y, A, x0, y0, sigma_x, sigma_y):
    return A * np.exp(-((x - x0)**2 / (2 * sigma_x**2) + (y - y0)**2 / (2 * sigma_y**2)))

# Parámetros de la función
A = 1      # Amplitud
x0 = 0     # Centro en x
y0 = 0     # Centro en y
sigma_x = 1  # Desviación estándar en x
sigma_y = 1  # Desviación estándar en y

# Crear una cuadrícula de valores para x e y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calcular la función gaussiana
z = gaussian_2d(x, y, A, x0, y0, sigma_x, sigma_y)

# Graficar la función gaussiana
plt.figure(figsize=(6, 5))
plt.contourf(x, y, z, cmap='viridis')
plt.colorbar(label='Amplitude')
plt.title('2D Gaussian Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

To convert this gaussian function into a likelihood, we must thinking as the probability the observing the data given the parameters(model).

A gaussian likelihood for errors for data generated by our baseline function (gaussian_2d), can be expreses as:

```python
from IPython.display import display, Math

display(Math(r'\mathcal{L}(\theta | \mathbf{d}) = \prod_{i} \frac{1}{\sqrt{2\pi} \sigma_i} \exp\left( - \frac{(d_{i} - f(x_{i}, y_{i}, \theta))^2}{2\sigma_{i}^2} \right)'))

```

Where $\( d_i = (x_i, y_i) \)$ represents the observed data, $\( f(x_{i}, y_{i}, \theta) \)$ is the Gaussian model generating the signal at the positions $\(x_i\) and \(y_i\)$, depending on the parameter set $\(\theta\)$, with $\(\theta = \{x_0, y_0, \sigma_x, \sigma_y, A\}\)$. Additionally, $\( \sigma_i \)$ denotes the variance of the noise for each measurement $\(i\)$.


Due to large amount of data, it's better to work with $log(\mathcal{L}(\mathbf{d} | \theta)$. Thus:
$$ \log \mathcal{L}(\mathbf{d} | \theta) = - \frac{1}{2} \sum_{i} \left[ \frac{(d_i - f(x_i, y_i, \theta))^2}{\sigma_i^2} \right]$$


## Defining the Gaussian likelihood and its parameters

```python
# Defining the likelihood function

import numpy as np
from scipy import stats

def my_lik(x, y, A, x0, y0, sigma_x, sigma_y):
    gaussian_value = gaussian_2d(x, y,A, x0, y0, sigma_x, sigma_y)
    return stats.norm.logpdf(gaussian_value, loc=1, scale=0.05)
# Dictionary for sampler


```
