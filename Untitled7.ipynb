```python
import healpy as hp
import matplotlib.pyplot as plt
import numpy as np
import pymaster as nmt
```

```python
# Creating new mask in equatorial coordinates
NSIDE = 64
ipix=hp.nside2npix(NSIDE)
m=np.zeros(ipix) #mask


ipix
```

```python
# Convertirg the the theta and phi for each pixels in decl and r.a.
for j in np.arange(1,ipix,1):
    theta, phi=hp.pix2ang(NSIDE, j)
    ra = np.rad2deg(phi)
    dec = np.rad2deg(0.5 * np.pi - theta)
    print(dec, ra)
```

```python
#convert the decl amongs -70 and 30 into pixels
for i in np.arange(-70,30,1):
    theta = 0.5 * np.pi - np.deg2rad(i)
    phi = np.deg2rad(ra)
    ipix_1=hp.ang2pix(NSIDE, theta, phi)
    print(ipix_1)
```
